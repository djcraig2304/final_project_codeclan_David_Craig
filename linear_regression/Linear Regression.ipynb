```python
import pandas as pd 
import numpy as np

```

```python
deloitte_kpis = pd.read_csv("clean_data/deloitte_gdp_kpis.csv")

```

```python
import pandas_profiling
```

```python
deloitte_kpis.head()
```

```python
ml_data = deloitte_kpis.loc[(deloitte_kpis["year"] >= 1995) & (deloitte_kpis["year"] < 2022) & (deloitte_kpis["developed"] == True)]

```

```python
ml_data.corr()
```

```python
import seaborn as sns
```

```python
all_data_ml = ml_data.drop(columns = ["mfp_annual_growth_rate_res", "gdp_hr_wrkd_usd", "gdp_per_capita_usd", 
                                                 "developed", "united_kingdom", "annual_gdp_growth", "date", "country_code",
                                     "labour_force_total", "expense_perc_gdp", "exchange_rate_lcu_usd", "oil_rents_perc_gdp", "poverty_headcount_ratio_perc_pop",
                                     "unemployment_perc_labour_force_total", "year"])

all_data_ml.corr()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
all_data_ml = all_data_ml.dropna() # dropped one na from 
```

```python
pandas_profiling.ProfileReport(all_data_ml)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
all_data_ml_dum = pd.get_dummies(all_data_ml, drop_first = True) # create dummy variables
```

```python
all_data_ml_dum.corr()
```

```python

y = all_data_ml_dum['gdp_usd']
X = all_data_ml_dum.drop(columns = 'gdp_usd')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_train, model.predict(X_train))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, model.predict(X_test))
```

```python
y_train_pred = model.predict(X_train)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y_train, y_train_pred)
plt.xlabel("Actual Target Variable")
plt.ylabel("Predicted Values")
plt.title("Multiple Linear Regression Model")
plt.savefig("linear_regression_plot.png")
```
